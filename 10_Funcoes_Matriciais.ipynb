### Modulo para imprimir arrays

Nessa aula iremos tratar com matrizes e arrays e ś sempre importante estar imprimindo eles para compreendermos o resultado. Para evitarmos a todo momento introduzirmos uma subrotina que imprima em cada um dos exemplos abaixo, vamos introduzirm um módulo que contém isso e vamos usá-lo no exemplos que se segue. O módulo é o seguinte:

````python
%%writefile src/Mod_ImpMat.f95
MODULE imprimir_matriz
   !---------------------------------------------------------------------
   IMPLICIT NONE
   !---------------------------------------------------------------------
   INTERFACE print_mat
      MODULE PROCEDURE print_mat_int
      MODULE PROCEDURE print_vec_int
      MODULE PROCEDURE print_mat_real
      MODULE PROCEDURE print_vec_real
   END INTERFACE
   !---------------------------------------------------------------------
CONTAINS
  !---------------------------------------------------------------------
  SUBROUTINE print_mat_int(b,dig,li,lc)
    !-------------------------------------------------------------------
    IMPLICIT NONE
    INTEGER, INTENT(in),       OPTIONAL :: dig, li, lc
    INTEGER, DIMENSION(:,:), INTENT(in) :: b
    CHARACTER(85)       :: fmt0, fmt1
    INTEGER :: i, j, n, m, ndig
    !-------------------------------------------------------------------
    IF ( PRESENT(li) ) THEN
       n = li
    ELSE
       n = SIZE(b,DIM=1)
    END IF
    IF ( PRESENT(lc) ) THEN
       m = lc
    ELSE
       m = SIZE(b,DIM=2)
    END IF
    IF ( PRESENT(dig) ) THEN
       ndig = dig
    ELSE
       ndig = NINT(LOG10(MAXVAL(REAL(abs(b)))))+ 1
    END IF
    WRITE(fmt0,'(i12)') ndig   ! Converte os INTEGER ndig no caracter fmt0
    fmt0 = '(I' // TRIM(ADJUSTL(fmt0)) // ",2x)"
    WRITE(fmt1,'(i12)') m   ! Converte os INTEGER m no caracter fmt0
    !-------------------------------------------------------------------
    fmt1 = '(1x,'// TRIM(ADJUSTL(fmt1)) // TRIM(ADJUSTL(fmt0)) // ')'
    !print *, 'fmt0 = ',fmt0, 'fmt1 = ', fmt1
    !-------------------------------------------------------------------
    DO i =1, n
       WRITE(*,fmt1)(b(i,j), j=1,m)
    END DO
    !-------------------------------------------------------------------
    RETURN
    !-------------------------------------------------------------------
  END SUBROUTINE print_mat_int
  !---------------------------------------------------------------------

  !---------------------------------------------------------------------
  SUBROUTINE print_vec_int(v,dig,ls)
    !-------------------------------------------------------------------
    IMPLICIT NONE
    INTEGER, INTENT(in),     OPTIONAL :: dig, ls
    INTEGER, DIMENSION(:), INTENT(in) :: v
    CHARACTER(85)       :: fmt0, fmt1
    INTEGER :: i, n, ndig
    !-------------------------------------------------------------------
    IF ( PRESENT(ls) ) THEN
       n = ls
    ELSE
       n = SIZE(v)
    END IF
    IF ( PRESENT(dig) ) THEN
       ndig = dig
    ELSE
       ndig = NINT(LOG10(MAXVAL(REAL(abs(v)))))+ 1
    END IF
    WRITE(fmt0,'(i12)') ndig   ! Converte os INTEGER ndig no caracter fmt0
    fmt0 = '(I' // TRIM(ADJUSTL(fmt0)) // ",2x)"
    WRITE(fmt1,'(i12)') n   ! Converte os INTEGER m no caracter fmt0
    !-------------------------------------------------------------------
    fmt1 = '(1x,'// TRIM(ADJUSTL(fmt1)) // TRIM(ADJUSTL(fmt0)) // ')'
    !-------------------------------------------------------------------
    WRITE(*,fmt1)(v(i), i=1,n)
    !-------------------------------------------------------------------
    RETURN
    !-------------------------------------------------------------------
  END SUBROUTINE print_vec_int
  !---------------------------------------------------------------------

  !---------------------------------------------------------------------
  SUBROUTINE print_mat_real(b,dig,li,lc)
    !-------------------------------------------------------------------
    IMPLICIT NONE
    INTEGER, INTENT(in),    OPTIONAL :: li, lc, dig
    REAL, DIMENSION(:,:), INTENT(in) :: b
    REAL    :: x
    INTEGER :: i, j, n, m, ndig, dec
    CHARACTER (LEN=95) :: fmt0, fmt1, fmt2
    !-------------------------------------------------------------------
    IF ( PRESENT(li) ) THEN
        n = li
    ELSE
        n = SIZE(b,DIM=1)
    END IF
    IF ( PRESENT(lc) ) THEN
        m = lc
    ELSE
        m = SIZE(b,DIM=2)
    END IF
    IF ( PRESENT(dig) ) THEN
       dec = dig
    ELSE
       dec = 2  ! Número de digitos padrão
    END IF
    x = MAXVAL(abs(b))
    IF ( x < 1.0 ) then
       ndig = NINT(LOG10(x)) + dec + 3
    ELSE
       ndig = NINT(LOG10(x)) + dec + 2
    END IF
    WRITE(fmt0,'(i12)') ndig     ! Numero total de digitos
    WRITE(fmt1,'(i12)') dec      ! Parte decimal
    fmt0 = '(F' // TRIM(ADJUSTL(fmt0)) // '.' // TRIM(ADJUSTL(fmt1)) // ',2x)'
    WRITE(fmt2,'(i12)') m        ! Numero de colunas
    fmt0 = '(1x,' // TRIM(ADJUSTL(fmt2)) // TRIM(ADJUSTL(fmt0)) // ')'
    !PRINT *, "fmt0 = ", TRIM(fmt0)
    !-------------------------------------------------------------------
    DO i =1, n
       WRITE(*,fmt0)(b(i,j), j=1,m)
    END DO
    !-------------------------------------------------------------------
    RETURN
    !-------------------------------------------------------------------
  END SUBROUTINE print_mat_real
  !---------------------------------------------------------------------

  !---------------------------------------------------------------------
  SUBROUTINE print_vec_real(v,dig,ls)
    !-------------------------------------------------------------------
    IMPLICIT NONE
    INTEGER, INTENT(in),  OPTIONAL :: ls, dig
    REAL, DIMENSION(:), INTENT(in) :: v
    REAL    :: x
    INTEGER :: i, n, ndig, dec
    CHARACTER (LEN=95) :: fmt0, fmt1, fmt2
    !-------------------------------------------------------------------
    IF ( PRESENT(ls) ) THEN
        n = ls
    ELSE
        n = SIZE(v)
    END IF
    IF ( PRESENT(dig) ) THEN
       dec = dig
    ELSE
       dec = 2  ! Número de digitos padrão
    END IF
    x = MAXVAL(abs(v))
    IF ( x < 1.0 ) then
       ndig = NINT(LOG10(x)) + dec + 3
    ELSE
       ndig = NINT(LOG10(x)) + dec + 2
    END IF
    WRITE(fmt0,'(i12)') ndig     ! Numero total de digitos
    WRITE(fmt1,'(i12)') dec      ! Parte decimal
    fmt0 = '(F' // TRIM(ADJUSTL(fmt0)) // '.' // TRIM(ADJUSTL(fmt1)) // ',2x)'
    WRITE(fmt2,'(i12)') n        ! Numero de colunas
    fmt0 = '(1x,' // TRIM(ADJUSTL(fmt2)) // TRIM(ADJUSTL(fmt0)) // ')'
    !PRINT *, "fmt0 = ", TRIM(fmt0)
    !-------------------------------------------------------------------
    WRITE(*,fmt0)(v(i), i=1,n)
    !-------------------------------------------------------------------
    RETURN
    !-------------------------------------------------------------------
  END SUBROUTINE print_vec_real
  !---------------------------------------------------------------------
END MODULE imprimir_matriz
````

E para usal o módulo basta introduzir a seguite instrução:

```fortran
PROGRAM test_mod_ImpMat
   USE imprimir_matriz
   ...
   CALL print_mat(Mat)      ! pode ser vetor ou matriz
   ...  
END PROGRAM test_mod_ImpMat  
```

Note que a interface

```fortran
   INTERFACE print_mat
      MODULE PROCEDURE print_mat_int
      MODULE PROCEDURE print_vec_int
      MODULE PROCEDURE print_mat_real
      MODULE PROCEDURE print_vec_real
   END INTERFACE
```

desse módulo, diz basicamamente que as 4 subrotinas `print_mat_int`, `print_vec_int`, `print_mat_real` e `print_vec_real` serão todas acessadas por um único nome `print_mat`. Ao chamar a função `print_mat` o Fortran irá definir que função ele irá chamar analisando os argumetos passados pela chamada da subrotina. Note portanto, que ao definir essa interface, as subrotinas que a compõem devem se diferenciar por pelo menos um de seus argumentos. 

Nos próximos programas iremos usar esse módulo

# Argumento Opcional MASK


Algumas funções matriciais como `MAXVAL`, `MINVAL`, `PRODUCT` e `SUM` tem um terceiro argumento opcional, uma matriz lógica `MASK`. Se este argumento estiver presente, ele deverá ter a mesma forma da matriz do primeiro argumento e a operação é aplicada somente aos elementos da matriz correspondendo aos elementos verdadeiros do argumento da matriz `MASK`; por exemplo, `SUM(A, MASK=A>0)` soma todos os elementos positivos da matriz `A`. O argumento `MASK` afeta somente os valores da função e não afeta o cálculo dos argumentos que são expressões matriciais.

## Funções matriciais: Exemplos

Três estudantes fazem 4 exames. Os resultados são armazenados em uma matriz inteira Nota(1:3,1:4) :

$$\text{nota}=\left[
\begin{array}{cccc}
85 & 76 & 90 & 60 \\
71 & 45 & 50 & 80 \\
66 & 45 & 21 & 55
\end{array}
\right]$$

- Maior Nota: 

```fortran
MAXVAL(nota) = 90
```

- Maior Nota de cada estudante: 

```fortran
MAXVAL(nota, DIM = 2) = (/ 90, 80, 66 /)
```

- Estudante com a maior nota: 

```fortran
MAXLOC(MAXVAL (nota, DIM = 2))    != MAXLOC((/ 90, 80, 66 /)) = (/ 1 /)
```

- Nota média: 

```fortran
media = SUM (nota) / SIZE (nota) ! = 62
```

Note que  a media é uma variavel `INTEGER`

- Número de notas acima da média: 

  ```fortran
  acima = nota > media
  ! acima(3, 4) eh uma matriz LOGICAL
  !         T T T F
  ! acima = T F F T
  !         T F F F
  n_maior_media = COUNT (acima) ! = 6
  ! n_maior_media eh uma variavel INTEGER
  ```

- Junte todas as notas acima da média : 

```fortran
...
INTEGER, ALLOCATABLE, DIMENSION (:):: &
pontos_maiores_media
...
ALLOCATE (pontos_maiores_media(n_maior_media))
pontos_maiores_media = PACK (nota, acima)      ! = (/ 85, 71, 66, 76, 90, 80 /)
```



- Todo e qualquer estudante está sempre acima da média?

  ```fortran
  ANY (ALL (acima, DIM = 2)) ! = .FALSE.
  ```

  

- Todos os estudante tiraram nos 4 exames pelo menos uma nota acima da média

  ```fortran
  ANY (ALL (acima, DIM = 1)) ! = .TRUE.
  ```

  ## Operações Matriciais

- As matrizes em todas as operações matriciais devem ter a mesma forma.
- Os cálculo são realizados elemento por elemento, isto é, as expressões são calculadas e somente depois o resultado é transferido.
- As funções podem ser calculadas matricialmente.
- Quando uma função implícita tem como argumento uma matriz, ela cálcula o valor da função implícita para cada elemento da matriz.

-  **Exemplo 1:**  

  - Versão Fortran 77:

  ```fortran
  REAL a(20), b(20), c(20)
     ...
  DO 10 i = 1, 20
     a(i) = 0.0
  10 CONTINUE
     ...
  DO 20 i = 1, 20
    a(i)=a(i)/3.1+b(i)*SQRT(c(i))
  20 CONTINUE
  ```

  

  - Versão Fortran 90:

  ```fortran
  REAL, DIMENSION(20):: a, b, c
  ...
  ! Zera os elementos de a
  a = 0.0
  ...
  ! Elemento por elemento
  ! recebe novo valor
  a = a/3.1 + b*SQRT(c)
  ```

  - Versão Fortran 77:

  ```fortran
  REAL a(5,5),b(5,5)
  REAL c(5,5),d(5,5)
  ...
     DO 20 i = 1, 5
        DO 10 j = 1, 5
           c(j,i)=a(j,i)*b(j,i)
           d(j,i)=sin(c(j,i))
  10    CONTINUE
  20 CONTINUE
  ```

  - Versão Fortran 90:

  ```fortran
  REAL, DIMENSION (5,5) :: a, b
  REAL, DIMENSION (5,5) :: c, d
  ...
  c = a * b
  d = sin(c)
  ...
  ```

- **Exemplo 2:** Situações específicos

Determine o valor máximo que seja menor que 1000 na matriz de três dimensões:

- O Fortran 77 necessita de um triplo loop `DO` e de condições `IF`.

- Fortran 90:

  ```fortran
  REAL, DIMENSION (10,10,10) :: a
  amax = MAXVAL(a, MASK = (a < 1000))
  ```

 
Determine o valor médio dos valores maiores que 3000 em uma matriz

- O Fortran 77 necessita de ninhos de  `DO` e de condições `IF`..

- Fortran 90:

  ```fortran
  av = SUM(a, MASK = (a > 3000)) / &
  COUNT(MASK=(a > 3000))
  ```

## A função  WHERE

- Forma:

  ```fortran
  WHERE (expresso_matricial_logica) atribuicao_matricial
  ```

  

- Se a expressão expresso\_matricial\_logica lógica for verdadeira, o comando atribuicao\_matricial será executado

  ```fortran
  REAL DIMENSION (5, 5) :: ra, rb
  ...
  WHERE (rb > 0.0) ra = ra / rb
  ```

- Note que a `MASK (rb >0.0)` deve ter a mesma forma dos índices de `ra` 

- Forma:

  ```fortran
  WHERE (expresso_matricial_logica)
    atribuicao_matricial
  END WHERE
  
  WHERE (expresso_matricial_logica)
   atribuicao_matricial
  ELSE WHERE (cond1)
   atribuicao_matricial
  ELSE WHERE (cond2)
   atribuicao_matricial
  ...
  ELSE WHERE (condn)
   atribuicao_matricial
  END WHERE
  ```

- **Exemplo 1:** Uma construção `WHERE` com duas possibilidades


```fortran
REAL DIMENSION (5, 5) :: ra, rb
...
WHERE (rb > 0.0)
   ra = ra / rb
ELSE WHERE
   ra = 0.0
END WHERE
```

- **Exemplo 2:** mais de uma condição


```fortran
PROGRAM Tabuleiro
  IMPLICIT NONE
  INTEGER, PARAMETER :: n = 10
  INTEGER :: i, j, diag
  INTEGER, DIMENSION(n,n) :: Board
  REAL, DIMENSION(n,n) :: mat

  CALL RANDOM_NUMBER(mat)
  mat = ANINT(100*mat)/10
  CALL imprimir(mat)
  WHERE ( Mat < 2 )
     mat = 1.0
  ELSEWHERE(Mat >= 2 .AND. Mat <= 5)
     mat = 4.0
  ELSEWHERE( Mat > 5 .AND. mat <= 7.5)
     mat =7.0
  ELSEWHERE (Mat > 7.5 .AND. Mat <= 10.0)
     mat = 9.0
  END WHERE
  CALL imprimir(mat)
  STOP
CONTAINS
  SUBROUTINE Imprimir(Mat)
    IMPLICIT NONE
    REAL, DIMENSION(:,:), INTENT(in) :: Mat
    !INTEGER :: DIM(2)
    INTEGER ::  i, j, nl, nc
    CHARACTER (LEN=85) :: fmt0
    !DIM = SHAPE(Mat)
    nl = SIZE(Mat,DIM=1)  ! nl = DIM(1);
    nc = SIZE(Mat,DIM=2)  ! nc = DIM(2)
    WRITE(fmt0,'(A4,I1,A10)') "(1X,",nc,"(f4.1,2x))"
    PRINT *, ""
    DO i = 1, nl
       WRITE(*,fmt0)(Mat(i,j), j=1,nc)
    END DO
    PRINT *, ""
    RETURN
  END SUBROUTINE Imprimir
END PROGRAM Tabuleiro
```

## Seções Matriciais 

Uma sub-matriz (chamada de seção), de uma matriz pode ser referenciada especificando um intervalo de sub-índices

- Um sub-índice simples

  ```fortran
  a(2, 3, 1) ! somente um elemento da matriz
  ```

  

- Um sub-índice triplo

  - `[índice inferior]:[índice superior][:passo]`
  - Quando omitidos, o índice inferior é 1 e o passo também é 1

- Um sub-índice vetorial

- **Exemplo**

```fortran
REAL, DIMENSION (5, 5) :: ra


Elemento matricial = ra(2:2:1, 2:2:1) ! shape (/ 1 /)


sublinha = ra(3, 3:5) ! Shape (/ 3 /)
```

### Sub-índices Vetoriais

- Uma matriz inteira 1D usada como sub-índice

```fortran
    (/ 3, 2, 12, 2, 1 /)
```

- Exemplo: 

```fortran
REAL, DIMENSION :: ra(6), rb(3)
INTEGER, DIMENSION (3) :: iv
iv=(/1, 3, 5/)       ! Expressao inteira de rank 1
ra=(/ 1.2, 3.4, 3.0, 11.2, 1.0, 3.7 /)
rb=ra(iv)            ! iv eh o indice do vetor
! = (/ ra(1), ra(3), ra(5) /)
! = (/ 1.2, 3.0, 1.0 /)
...
```

### Sub-índices Vetoriais 

- Sub-índices vetoriais podem estar no lugar das expressões dos limites dos sub-índices

```fortran
iv = (/ 1, 3, 5 /)
ra(iv) = (/ 1.2, 3.4, 5.6 /)
! =ra((/ 1, 3, 5 /))=(/ 1.2, 3.4, 5.6 /)
```

- Não se pode repetir valores no vetor dos sub-índices

```fortran
iv     = (/ 1, 3, 1 /)       ! Permitido
ra(iv) = (/ 1.2, 3.4, 5.6 /) ! Proibido
! = ra((/ 1, 3, 1 /)) = (/ 1.2, 3.4, 5.6 /)

iv     = (/ 1, 3, 5 /)
ra(iv) = (/ 1.2, 3.4, 5.6 /) ! Permitido
```

### Alocação Matricial

- As matrizes das operações devem possuir a mesma forma. Exemplo: 

```fortran
REAL, DIMENSION (5, 5) :: ra, rb, rc
INTEGER :: id
...
ra           = rb + rc * id                    ! Shape(/ 5, 5 /)
ra(3:5, 3:4) = rb(1::2, 3:5:2) + rc(1:3, 1:2)  ! Shape(/ 3, 2 /)
ra(:, 1)     = rb(:, 1) + rb(:, 2) + rb(:, 3)  ! Shape(/ 5 /)
```

### Recorrência Reversa 

O seguinte código:

```fortran
    DO i=2,n
       x(i) = x(i) + x(i-1)         !  (1)
    END DO
```

não é o mesmo que:

```fortran
       x(2:n) = x(2:n) + x(1:n-1)   !  (2)
```

- Em (1):

```fortran
    x(i) = x(i) + x(i-1) + x(i-2) + ... + x(1)
```

- Equivalente em Fortran 90:

```fortran
    x(2:n) = (/ (SUM(x(1:i)), i = 2, n) /)
```


### Locação Elementar versus Sub-índice

`MAXLOC`/`MINLOC` retorna o elemento max/min localizado em vez do sub-índice de sua localização

```fortran
REAL,  DIMENSION(1:8) :: ra
REAL,  DIMENSION(3:4) :: rb
INTEGER, DIMENSION(1) :: locmax1, locmax2
REAL :: max1, max2
ra=(/1.2, 3.4, 5.4, 11.2, 1.0, 3.7, 1.0, 1.0/);  rb = ra
! Para encotrar a localizacao do valor max.
locmax1 = MAXLOC(ra) ! = (/ 4 /)
locmax2 = MAXLOC(rb) ! = (/ 4 /)
! Para encotrar o valor max. a partir da localizacao
max1 = ra(locmax1(1))
! Devido a ra ser definido com um indice inferior menor que 1
max2 = rb(LBOUND(rb) + locmax2(1) - 1)
! Forma geral necessaria quando indice inferior eh menor que 1
```

### Matrizes Com Dimensão Excedida

- Quando o limite inferior excede o limite superior, a matriz tem um tamanho zero
- São válidas e seguem as regras normais
- Em operações devem possuir a mesma forma matricial (se uma forma de tamanho zerado, todas envolvidas na operação também devem ser) 
- Usuais nas fronteiras: Os códigos não precisam tratar a fronteira de forma especial


#### Exemplo

```fortran

  x(i)    = b(i) / a(i, i)
  b(i+1:n)= b(i+1:n)-a(i+1:n,i)*x(i)
```

## Algoritmo dos gradientes conjugados


```fortran
PROGRAM gradientes_conjugados
IMPLICIT NONE
INTEGER :: iters, its, n
LOGICAL :: convergido
REAL :: tol, acima, alfa, beta
REAL, ALLOCATABLE :: a(:,:),b(:),x(:),r(:),u(:),p(:),xnovo(:)
READ (*,*) n, tol, its
ALLOCATE(a(n,n),b(n),x(n),r(n),u(n),p(n),xnovo(n))
OPEN (10, FILE='dados.dat')
READ (10,*) a, b
x = 1.0;       r = b - MATMUL(a,x)
p = r;     iters = 0
DO
  iters = iters + 1;    u = MATMUL(a, p)
  acima = DOT_PRODUCT(r, r)
  alfa  = acima / DOT_PRODUCT(p, u)
  xnovo = x + p * alfa;     r = r - u * alfa
  beta = DOT_PRODUCT(r, r) / acima
  p = r + p * beta
  convergido = (MAXVAL(ABS(xnovo - x))/MAXVAL(ABS(x)) < tol )
  x = xnovo
  IF (convergido .OR. iters == its) EXIT
END DO
WRITE (*,*) "iteracoes = ", iters, " x = ", x
END PROGRAM gradientes_conjugados
```

## Matrizes multidimensionais e seu armazenamento interno 

Há algumas situações que é necessário saber como o Fortran armazena suas matrizes na memória, e esse é o caso quando desejameo escrever um código otimizado ou quando se desejar misturar códigos escritos na liguagem C com códigos escritos em Fortran. Portanto considere a seguinte matriz bidimesnional:
$$
M = \begin{pmatrix}
M_{1,1} &  M_{1,2} &  M_{1,3} & M_{1,4} \\
M_{2,1} &  M_{2,2} &  M_{2,3} & M_{2,4} \\
M_{3,1} &  M_{3,2} &  M_{3,3} & M_{3,4} \\
M_{4,1} &  M_{4,2} &  M_{4,3} & M_{4,4}
\end{pmatrix}
$$

-  <span style="color:red; font-size:1.2em">Na linguagem C</span> essa matriz será armazenada na memória da seguinte forma:

$$
\begin{array}{ccccccccc}
\text{Início} & M(1,1) & \longrightarrow  & M(1,2) & \longrightarrow  & M(1,3) & \longrightarrow  & M(1,4) & \downarrow \\ 
\downarrow  & \longleftarrow & \longleftarrow  & \longleftarrow & \longleftarrow  & \longleftarrow & \longleftarrow   & \longleftarrow & \hookleftarrow \\
\hookrightarrow & M(2,1) & \longrightarrow  & M(2,2) & \longrightarrow  & M(2,3) & \longrightarrow  & M(2,4) & \downarrow  \\ 
\downarrow  & \longleftarrow & \longleftarrow  & \longleftarrow & \longleftarrow  & \longleftarrow & \longleftarrow   & \longleftarrow & \hookleftarrow \\ 
\hookrightarrow & M(3,1) & \longrightarrow  & M(3,2) & \longrightarrow  & M(3,3) & \longrightarrow  & M(3,4) & \downarrow \\
\downarrow  & \longleftarrow & \longleftarrow  & \longleftarrow & \longleftarrow  & \longleftarrow & \longleftarrow   & \longleftarrow & \hookleftarrow \\ 
\hookrightarrow & M(4,1) & \longrightarrow  & M(4,2) & \longrightarrow  & M(4,3) & \longrightarrow  & M(4,4) &  \text{Fim}\\
\end{array}
$$

-  <span style="color:red; font-size:1.2em">Na liguagem Fortran</span> essa matriz será armazenada na memória da seguinte forma:

$$
\begin{array}{EFEFEFEF} %{lclclcl}
\text{Início} & \Rsh  & \downarrow  & \Rsh  & \downarrow  & \Rsh & \downarrow  \\
\downarrow  & \uparrow  & \downarrow  & \uparrow  & \downarrow  & \uparrow & \downarrow  \\
M(1,1) & \uparrow  & M(1,2) & \uparrow  & M(1,3) & \uparrow  & M(1,4) \\
M(2,1) & \uparrow  & M(2,2) & \uparrow  & M(2,3) & \uparrow  & M(2,4) \\
M(3,1) & \uparrow  & M(3,2) & \uparrow  & M(3,3) & \uparrow  & M(3,4) \\
M(4,1) & \uparrow  & M(4,2) & \uparrow  & M(4,3) & \uparrow  & M(4,4) \\
\downarrow  & \uparrow  & \downarrow  & \uparrow  & \downarrow  & \uparrow & \downarrow  \\
\hookrightarrow  & \uparrow  & \hookrightarrow  & \uparrow  &
\hookrightarrow  & \uparrow  & \text{Fim}\\
\end{array}
$$

Portanto, para obter um código mais eficiente em Fortran, ao acessar essa matriz, deve-se usar as fatias da seguinte forma `M(:, 1)`, `M(:, 2)`, ou no caso de uma matriz `M` com uma dimensão a mais `M(:, :, 1)`, por exemplo. os dois pontos devem estar à esquerda. Dessa forma, o passo será contínuo e mais rápido. Então, quando precisa de alguma fatia de uma matriz no seu algoritmo, sempre configure a matriz de forma, para que seja possível chamar ela como foi mstrado acima. Se colocar um `:` à direita, como `M(1,:)`, o seu código será mais lento.

#### Exemplo:

```fortran
RES = matmul(A(:, :, i), B)         ! Rápido 
RES = matmul(A(i, :, :), B)         ! Lento
```

Em outras palavras, a *ordem de armazenamento do Fortran é*: índice de loop menor /mais rápido /mais interno primeiro, maior/mais lento/ último índice de loop externo, ou seja, "*O mais interno é o mais à esquerda*". Assim, os elementos de um array 3D `Mat(N1, N2, N3)` são armazenados e, portanto, acessados com mais eficiência, como:

```fortran
do i3 = 1, N3
    do i2 = 1, N2
        do i1 = 1, N1
            Mat(i1, i2, i3) = ...
        end do
    end do
end do
```

A matriz associada de vetores seria então acessada com mais eficiência como:

```fortran
do i3 = 1, N3
    do i2 = 1, N2
        Mat(:, i2, i3) = ...
    end do
end do
```

E o conjunto associado de matrizes seria acessado com mais eficiência como:

```fortran
do i3 = 1, N3
    Mat(:, :, i3) = ...
end do
```

Para armazenar/acessar com eficiência como acima, então sempre acesse os blocos de memória  contínua, diretamente adjacentes uns aos outros; sem pulos/passos.

Quando não se tem certeza, sempre reescreva (na sua cabeça) o algoritmo para usar pedaços, por exemplo, o primeiro loop se tornaria:

```fortran
do i3 = 1, N3
    Mat_i3 = Mat(:, :, i3)
    do i2 = 1, N2
        Mat_i2i3 = Mat_i3(:, i2)
        do i1 = 1, N1
            Mat_i2i3(i1) = ...
        end do
    end do
end do
```

o segundo loop se tornaria:

```fortran
do i3 = 1, N3
    Mat_i3 = Mat(:, :, i3)
    do i2 = 1, N2
        Mat_i3(:, i2) = ...
    end do
end do

```

E certifique-se de que todos os passos estão sempre à esquerda. Então, dessa forma o seu código será rápido. Tudo o que foi aprsentado aqui também vale para a linguagem C, entretanto na ordem inversa, por isso, para escrevermos códigos que envolvem cálculos ou manipulações matriciais eficientes devemos conhecer como essas matrizes são armazenadas na memória.

# Outros trechos de códigos

## Especificações

<span style="color:red">tipo [[,DIMENSION (dimensoes)] [,atributos]... ] :: lista_de_variaveis</span>

- **Tipo** <span style="color:red">INTRINSIC</span> ou tipo derivado;
- **DIMENSION** Opcional, mas necessita que as dimensões padrões sejam definidas;
- **(dimensoes)** Fornece as dimensões da matriz:
   - Constante inteira;
   - Expressão inteira usando argumentos mudos ou constantes;
   - Se uma matriz está alocada ou com uma forma assumida.  
- **atributos** Os mesmos vistos anteriormente;
- **lista_de_variaveis** lista dos nomes das matrizes, opcionalmente com as dimensões e os valores iniciais.

A seguir apresentamos alguns exemplos de declarações:

- Matrizes bidimensionais
```Fortran
REAL, DIMENSION(-3:4, 7) :: ra, rb
```

- Definindo as dimensões e inicializando:
```Fortran
INTEGER, DIMENSION(3):: a = (/ 1, 2, 3 /), b = (/ (i, i = 15, 17) /)
```

- Matrizes Automáticas:
```Fortran
LOGICAL, DIMENSION(8)         :: la
LOGICAL, DIMENSION (SIZE(la)) :: lb
```
na qual <span style="color:red">la</span> é um argumento matricial mudo.

- Estas são todas as formas matriciais explicitas.

- A seguir apresentamos um exemplo com alguns tipos de declarações

```Fortran
REAL :: x, y
INTEGER, PARAMETER :: n=5
REAL, DIMENSION (n,n) :: a, b(n), c(n+1,n+2), m
REAL :: e(n,n), f(n+2)
...
a = 3.0                      ! Todos os elementos de a recebem 3
f = (/ (2*i, i=n+1,2*n+3) /) ! do implicito
b = (/ 1, 2, 3, 4, 5 /)      ! é um vetor
c = (/ (i, i=1,42) /)        ! do implicito
FORALL(i=1:n, j=1:n)
   m(i,j) = REAL(i+j)/REAL(J)
END FORALL
e = 2.0 * m * exp( -m )      ! Calculo matricial elemento por elemento
...
```

- A seguir mostramos um segundo exemplo alguns tipos de declarações

```Fortran
REAL ::P                ! x e p tem uma forma explicita
REAL ::x(10,1:5,-2:3)   ! os indices sao constantes
DIMENSION :: p(1500)    ! Declara tamanho de p
INTEGER :: ibuff(:,:), obuff (:) ! Matrizes de forma assumida
INTEGER :: ctes (mdim,ndim)      ! Matriz com uma forma explicita
COMPLEX, ALLOCATABLE, DIMENSION (:,:) :: coords ! Matriz de forma definida e dimensão a ser definida
REAL, POINTER :: ptr(:,:,:)   ! Ponteiro de rank 3, forma definida e dimensão a ser definida
CHARACTER*5 :: text(10,*)     ! Matriz caracter de tamanho assumido
```

- Matrizes Alocadas Dinamicamente:

```Fortran
REAL, DIMENSION(:,:), ALLOCATABLE :: a, b
```
As dimensões são definidas posteriormente com o comando <span style="color:red">ALLOCATE</span>.

- Matrizes de Forma Assumida:
```Fortran
SUBROUTINE test(np, a, b, c)
 IMPLICIT NONE
 INTEGER, INTENT(OUT) :: np
 REAL, DIMENSION(:, :, :), INTENT(IN) :: a, b
 REAL, DIMENSION(:,:),ALLOCATABLE,INTENT(OUT)::c
 ...
END SUBROUTINE test
```
As dimensões são passadas como argumentos dos subprogramas.

- Matrizes de Forma Assumida, considere o seguinte trecho  de código:

```Fortran
 IMPLICIT NONE
 ...
 REAL, DIMENSION(-4:12, -4:12) :: MatA
 REAL, DIMENSION(-8:8, -8:8)   :: MatB
 ...
 CALL test(linfa,linfB,MatA,MatB,MatC) 
```

```Fortran
SUBROUTINE test(linfa,linfB,A,B,C)
 IMPLICIT NONE
 INTEGER, INTENT(in) :: linfa, linfb
 REAL, DIMENSION(linfa:, linfa:), INTENT(IN) :: a
 REAL, DIMENSION(linfb:, linfb:), INTENT(IN) :: b
 REAL, DIMENSION(:,:),ALLOCATABLE,INTENT(OUT)::c
 ...
END SUBROUTINE test
```

O tamanho das dimensões são passadas como argumentos dos subprogramas, entretato não o seu limite inferior, por isso ele deve ser enviado explicitamente.
