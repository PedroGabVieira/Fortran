Os *arrays* ou matrizes são um conjunto de dados, todos do mesmo tipo,  ordenados de alguma forma. No Fortran, eles podem ser de qualquer um dos tipos intrínsecos ou ainda de um tipo derivado.

No caso de matrizes no Fortran 90/95, há três considerações a serem feitas:

-  Elas nos permitem referir a um conjunto ou grupo de itens por um único nome.
-  Elas nos permitem referir a itens individuais ou membros desse conjunto, ou seja, procurá-los.
-  É possível escolher uma estrutura de controle que permita fácil manipulação deste conjunto ou array.

Um recurso muito importante para a comunidade científica introduzido no Fortran 90 foram os recursos para processamento de matrizes. Com eles, tornou-se possível trabalhar diretamente com uma matriz inteira ou uma seção de matriz sem a necessidade de loops `DO` explícitos. As funções intrínsecas agora podem agir de forma elementar nas matrizes e as funções podem ter valores de matriz. Também estão disponíveis as possibilidades de matrizes alocáveis, matrizes de formas assumidas e matrizes dinâmicas.  Além da operações matriciais e várias funções matriciais intrínsecas introduzidas pelo Fortran 90, ele também  introduziu o conceito de funções que retornam matrizes. Tais funções podem ser usadas para um estilo de programação muito compacto e de alto nível.

Enquanto Fortran 90 permitia ao programador definir funções que retornavam matrizes de dados, o Fortran 95 introduziu o conceito de funções e procedimentos elementares. Em combinação com vários outros recursos novos que foram introduzidos no Fortran 2003, estes se tornaram ainda mais poderosos e fáceis de usar.

Muitas das funções intrínsecas funcionam diretamente em arrays e as operações de matriz do Fortran transformam um array em outro, sem a necessidade de loops explícitos que possam obscurecer o objetivo do programador. Por exemplo, suponha que você tenha uma matriz de dados e deseje imprimir apenas os valores maiores que algum `limite`. Uma maneira de fazer isso é:

```fortran
do i = 1,size(vec)
	if ( vec(i) > limite ) then
		write(*,'(f10.4)') vec(i)
	end if
end do
```

É possível usando a função intrínseca `PACK()` (a qual detalharemos posteriormente), reescrever essa estrutura na seguinte forma mais compacta:

```fortran
write(*,'(f10.4)') pack( vec, vec > limite )
```

Quanto as matrizes, no Fortran 90/95 há 3 tipos possíveis, dependendo da relação da matriz com a quantidade de memória disponível para o armazenamento, elas podem ser classificadas como:

1. **Matrizes estáticas:** seu tamanho é fixo quando o array é declarado e não pode ser alterado durante a execução. Isso é inflexível em certas circunstâncias (é necessário editar o programa, alterando a(s) dimensão(ões) e recompilar) há desperdício de memória em termos de espaço de armazenamento (já que é possível declarar uma matriz muito maior do que o necessário para evitar de editar o código como no problema anterior).
2. **Matrizes semi-dinâmicas:** o tamanho de uma matriz é determinado após a entrada em uma subrotina e as matrizes podem ser criadas para corresponder ao tamanho exato requerido, mas só podem ser usadas ao serem passadas para a subrotina. Em Fortran90 esses arrays são chamados de matriz de forma presumida e automática ou assumida e automática.
3. **Matrizes dinâmicas:** o tamanho e, portanto, a quantidade de armazenamento usada por um *array* dinâmico pode ser alterada durante a execução. Isso é muito flexível, mas como efeito colateral possui um desempenho lento em tempo de execução e  falta de qualquer verificação de limite durante a compilação pode levar a um estouro de memória da máquina, ao tentar alocar um array maior do que a memória disponível no mmento. Em Fortran90, esses arrays são chamados de *arrays* de alocação dinâmica.

As razões para usar uma matriz são:

-  Mais fácil de declarar (um nome de variável em vez de dezenas ou até milhares).
-  Mais fácil de operar (devido às operações de matriz inteira, o código está mais próximo da forma matemática subjacente).
-  Acesso flexível (pode-se operar facilmente em várias áreas de matrizes de diferentes maneiras).
-  Mais fácil de entender o código (conveniência notacional).
-  Paralelismo de Dados Inerentes (execute um cálculo semelhante em muitos objetos de dados simultaneamente).
-  Oportunidades de otimização (para designers de compiladores).
-  Redução do tamanho do programa.

Esses e outros novos recursos serão descritos aqui, mas antes de tudo é necessário introduzir algumas terminologias.

## Terminologia

Fortran permite que um array tenha até sete subscritos, cada um relacionado a uma dimensão do array. As dimensões de uma matriz podem ser especificadas usando um atributo de dimensão ou uma especificação de matriz. Por padrão, as dimensões da matriz começam em 1, mas um intervalo diferente de valores pode ser especificado fornecendo um limite inferior e um limite superior. Por exemplo,

```fortran
REAL, DIMENSION(50) :: w      ! w(1), w(2), ..., w(50)
REAL, DIMENSION(5:54) :: x    ! x(5), x(6), ..., x(54)
REAL y(50)                    ! y(1), y(2), ..., y(50)
REAL z(11:60)                 ! z(11), z(12), ..., z(60)
```

-  Aqui, `w`, `x`, `y` e `z` são todos os arrays contendo 50 elementos.
-  A classificação de uma matriz é o número de dimensões. Assim, um escalar tem posto ou *rank* 0, um vetor tem posto 1 e uma matriz tem posto 2.
-  A dimensão/extensão (*extent*) refere-se a uma dimensão específica e é o número de elementos nessa dimensão.
-  A forma (*shape*) de uma matriz é um vetor que consiste na extensão de cada dimensão.
-  O tamanho (*size*) de uma matriz é o número total de elementos que compõem a matriz. Isso pode ser zero.
-  Duas matrizes são consideradas conformáveis (*conformance*) se tiverem a mesma forma. Todas as matrizes são compatíveis com um escalar.

Antes de continuarmos vamos deixar  claro o significado no Fortran de cada um dos termos abaixo:

-  `Rank` :   Número de dimensões;
-  `Extent` :   Número de elementos de uma das dimensões da matriz;
-  `Shape` :   Vetor que contém o número de elementos de cada uma das dimensões da matriz;
-  `Size` :   Produto do número de elementos de cada umas das dimensões da matriz;
-  `Conformance` :   O mesmo que `shape`

Veja, por exemplo, os seguintes arrays:

```fortran
REAL, DIMENSION :: a(-2:5,7)
REAL, DIMENSION :: b(8,-1:5)
REAL, DIMENSION :: d(11:18,1:8)
INTEGER :: c
```

A matriz `a` tem 

-  `Rank` 2
-  `Extent` 8 e 7
-  `Shape`  (/8,7/)
-  `Size`  56

Além disso, `a` está em  *conformance* com `b` e `c`, como `b` tem forma `(/ 8,7 /)` e `c` é escalar. No entanto, `a` não está em *conformance* com `d`, já que  `d` tem forma `(/ 8,9 /)`.

## Especificação

A forma geral de uma especificação de matriz é a seguinte:

```fortran
TYPE [[,DIMENSION (dimesões)] [,atributo]... ::] lista_de_variaveis
```

para a qual 

-  `TYPE`:  ele pode ser um tipo `INTRINSIC` ou um tipo derivado;
-  `DIMENSION` : É opcional, mas necessita que as dimensões padrões sejam definidas;
-  `(dimensões)`:   Fornece as dimensões da matriz, as quais devem ser especificadas por
   -  Constantes inteiras;
   -  Expressões inteiras usando argumentos mudos ou constantes;
   -  `:` é usado para mostrar que a matriz é uma forma alocável ou assumida
-  `atributo`:  esse parãmetro pode ser qualquer um dos seguintes:
   -  `PARAMETER`
   -  `PUBLIC`
   -  `PRIVATE`
   -  `POINTER`
   -  `TARGET`
   -  `ALLOCATABLE`
   -  `DIMENSION(dimesões)`
   -  `INTENT(inout)`
   -  `OPTIONAL`
   -  `SAVE`
   -  `EXTERNAL`
   -  `INTRINSIC`
-  `lista_de_variaveis`:   lista dos nomes das matrizes, opcionalmente com as dimensões e os seus valores iniciais.

Note que:

-  Aqui, `TYPE` pode ser qualquer tipo intrínseco ou um tipo derivado (desde que a definição de tipo derivado seja acessível para a unidade de programa que declara a matriz). 
-  `DIMENSION` é opcional e define as dimensões padrão da lista `dimesões`, que podem ser alternativamente definidas na lista `lista_de_variaveis`.

Os exemplos a seguir mostram a forma da declaração de vários tipos de matrizes, alguns dos quais são novos no Fortran 90 e serão encontrados mais tarde aqui:

-  Declaração de matrizes com duas dimensões:

   ```fortran
   REAL, DIMENSION(-3:4, 7) :: mat_a, mat_b
   ```

-  Inicialização de matrizes unidimensionais contendo 3 elementos:

```fortran
INTEGER, DIMENSION(3) ::  mat_a = (/1,2,3/),  & 
                          mat_b = (/ (i, i=15,19,2) /)
```

No padrão fortran 2003 temos

```fortran
INTEGER, DIMENSION(3) ::  mat_a = [1,2,3],  & 
                          mat_b = [(i, i=15,19,2)]
```

-  Declaração de matriz automática. Aqui `mat_a` é um argumento matricial mudo, e `SIZE` é uma função intrínseca que retorna um inteiro padrão escalar correspondente ao tamanho do `mat_a` da matriz:

```fortran
REAL, DIMENSION(SIZE(mat_a)) :: mat_b
```

-  Declaração de matrizes dinâmicas (atribuíveis) `mat_a` e `mat_b`. As dimensões seriam definidas em uma instrução `ALLOCATE` subsequente:

```fortran
REAL, DIMENSION (:,:), ALLOCATABLE :: mat_a, mat_b
```

-  Declaração de matrizes de formas assumidas `mat_a` e `mat_b`. As dimensões seriam tiradas da rotina de chamada real:

```fortran
REAL, DIMENSION(:,:,:) :: mat_a, mat_b
```

Estas são todas as formas matriciais explicitas. A seguir apresentamos mais alguns exemplos.

### Exemplo 1 

No exemplo abaixo, a forma de atribuição implícita é válida a partir do Fortran 90.

```fortran
REAL :: x, y
REAL, DIMENSION (5,5) :: a, b(5),  c(6,7), m
REAL :: e(5,5), f(7)     ! o atributo DIMENSION é opcional
...
! A forma de atribuição a seguir é válida a partir do Fortran 90
a = 3.0  ! Todos os elementos de a recebem 3
!f = (/ (2*i, i=6,12) /) ! do implícito padrão Fortran 90/95
f =[ (2*i, i=6,12) ]     ! do implícito a partir do Fortran 2003
b = (/ 1, 2, 3, 4, 5 /) ! é um vetor - Válida para o fortran 90
c = RESHAPE((/ (i, i=1,42) /),(/ 6,7/))    ! do implícito

FORALL(i=1:5, j=1:5)
   m(i,j) = REAL(i+j)/REAL(J)
END FORALL
e = 2.0 * m * exp( -m ) ! Calculo matricial
                        ! elemento por elemento
...
```

## Operações sobre a matriz inteira

No Fortran 77, não era possível realizar uma operação sobre uma matrizes inteira, em vez disso, cada elemento da mariz tinha que ser trabalhado separadamente, muitas vezes exigindo o uso de loops-DO aninhados. Ao lidar com matrizes grandes, tais operações poderiam consumir muito tempo e, além disso, o código necessário era muito difícil de ler e interpretar. Um novo recurso importante do Fortran 90 é a capacidade de executar operações sobre matrizes inteiras, permitindo que uma matriz seja tratado como um único objeto escalar e eliminando a necessidade de loops complicados e ilegíveis.

Para que essas operações sobre uma matriz inteira sejam executadas, as matrizes envolvidas devem ter as mesmas dimensões, ou sejma devem estar em conformance. Lembre-se de que, para duas matrizes estejam em conformance, elas devem ter a mesma forma, e qualquer matriz é compatível com um escalar. As operações entre dois arranjos conformáveis são realizadas elemento por elemento, e todas as operações intrínsecas são definidas entre dois desses arrays.

Por exemplo, se as matrizes $A$ e $B$ forem ambas de dimensão $2\times 3$, como no exemplo  a seguir

\begin{equation}
A = \begin{pmatrix}
3 & 5 & 9 \\
1 & 7 & 2
\end{pmatrix}
\qquad\qquad \text{e} \qquad\qquad
B = \begin{pmatrix}
8 & 2 & 1 \\
5 & 6 & 4
\end{pmatrix}
\end{equation}

o resultado da adição e subtração entre elas é 

\begin{equation}
A + B = \begin{pmatrix}
11 & 7 & 10 \\
6 & 13 & 6
\end{pmatrix}
\qquad\qquad \text{e} \qquad\qquad
A - B = \begin{pmatrix}
-5 & 3 & 8 \\
-4 & 1 & -2
\end{pmatrix}
\end{equation}

Já o seu produto elemento por elemnto é
\begin{equation}
A * B = \begin{pmatrix}
24 & 10 & 9 \\
5 & 42 & 8
\end{pmatrix}
\end{equation}

Podemos ainda adicionar uma constante, por exmplo, 7 a cada elemento de $A$, logo teremos

\begin{equation}
A + 7 = \begin{pmatrix}
10 & 12 & 16 \\
8  & 14 & 9
\end{pmatrix}.
\end{equation}

Essa difusão de escalares é útil ao inicializar matrizes e reescalonar os valores de uma matrize, ou mesmo, normalizar um vetor, etc.

Todas as funções intrínsecas que atum em escalares podem ser usada para realizar, operações matrciais sobre toda a matriz, assim


\begin{equation}
A = \log(B) = 
\begin{pmatrix}
\log(8) & \log(2) & \log(1) \\
\log(5) & \log(6) & \log(4)
\end{pmatrix}
\qquad\qquad \text{ou} \qquad\qquad
A = \sqrt{B} = 
\begin{pmatrix}
\sqrt{8} & \sqrt{2} & \sqrt{1} \\
\sqrt{5} & \sqrt{6} & \sqrt{4}
\end{pmatrix}
\end{equation}

Um conceito importante em relação à atribuição com valor de matriz é que a avaliação do lado direito é calculada antes que qualquer atribuição ocorra. Isso é importante quando uma matriz aparece nos lados esquerdo e direito de uma atribuição. Se esse não for o caso, os elementos na matriz do lado direito poderão ser afetados antes que a operação seja concluída.

Essa operções simplificam a escrita e interpretação de um código, aproximando cada vez mais o seu código do **For**mula **Tran**slation.

## Matrizes de tamanho zero

Pode-se pensar que uma matriz sempre teria pelo menos um elemento. No entanto, esse requisito forçaria os programas a conter um código extra para lidar com certas situações naturais. Como exemplo vamos usar a solução de um sistema na forma triangular inferior, ou seja:

\begin{equation}
\begin{bmatrix}
l_{1,1} &   0    &   0    & \cdots & 0 \\
l_{2,1} & l_{2,2} &   0    & \cdots & 0 \\
l_{3,1} & l_{3,2} & l_{3,3} & \cdots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
l_{n,1} & l_{n,2} & l_{n,3} & \cdots & l_{n,n}
\end{bmatrix}
\begin{bmatrix}
x_{1} \\
x_{2} \\
x_{3} \\
\vdots \\
x_{n}
\end{bmatrix}
=
\begin{bmatrix}
c_{1}  \\
c_{2}  \\
c_{3}  \\
\vdots  \\
c_{n}
\end{bmatrix}
\end{equation}
Por substituições retroativas, a solução deste sistema pode ser representada por:

\begin{equation}
\boxed{
x_{i} = \dfrac{ c_{i} -  \sum\limits_{j=1}^{i-1} l_{i,j}
x_{j} }{l_{i,i}}
 , \quad i=1,2, \dotsc, n. }
\end{equation}

Para ficar mais clara vamos abrir os primeiros termos:

\begin{align*}
   x_1 & = \frac{c_1}{l_{1,1}} \\
   x_2 & = \frac{c_2 - l_{2,1}x_1}{l_{2,2}} \\
   x_3 & = \frac{c_3 - l_{3,1}x_1 -l_{3,2}x_2}{l_{3,3}} \\
   \vdots\quad  & = \qquad \qquad \vdots
\end{align*}

Essa expressão é apresentada no algoritmo abaixo, o qual é conhecido como versão coluna.

![sistema_triangular_inferior](./Figs/Algoritmo_Sis_Triang_Inf.png)

Entretanto há uma versao coluna, que pode ser obtida invertendo a ordem do loop, com isso obtemos o seguinte algoritmo:

![Sitema triangula inferior vesão coluna](./Figs/Algoritmo_Sis_Triang_Inf_col.png)

Note nesse caso, que ele altera o valor da variável $c$. 

Por exemplo, o trecho de código a seguir resolve um conjunto triangular superior de equações lineares. Quando o valor de `i` é igual a `n`, as seções matriciais de `c` têm tamanho zero, que é exatamente o que é necessário.

```fortran
! Um loop cuja iteração final tem uma matriz de tamanho zero.
do i = 1, n
   x(i) = c(i) / l(i, i)
   c(i + 1:n) = c(i + 1:n) - l(i + 1:n, i) * x(i)
end do
```

O Fortran permite que as matrizes tenham tamanho zero em todos os contextos. **Sempre que um limite inferior excede o limite superior correspondente, a matriz terá tamanho zero.**

Existem poucas regras especiais para matrizes de tamanho zero porque elas seguem as regras usuais, embora alguns cuidados possam ser necessários em sua interpretação. Por exemplo, duas matrizes de tamanho zero da mesma classificação podem ter formas diferentes. Um pode ter forma `(0,2)` e a outra `(0,3)` ou `(2,0)`.

Essas matrizes de formas diferentes não possuem a mesma conformance e, portanto, não podem ser usadas juntas como operandos de uma operação binária. No entanto, um array é sempre compatível com um escalar, de modo que a declaração é válida e o escalar é "*transmitido para todos os elementos da matriz*", tornando isso uma declaração do tipo "*faça nada*".

Uma matriz de tamanho zero é considerada como sendo sempre definida, porque não possui valores que possam ser indefinidos.

### Argumento opcional DIM

-   Todas as funções matriciais tem um segundo argumento opcional `DIM` que é
    um escalar inteiro.

-   Se `DIM` estiver presente então a operação é aplicada a todas as
    sub-matrizes de `rank` 1 que se estendem através das outras
    dimensões `DIM` para produzir uma matriz de `rank` reduzido por 1 e de
    extensão igual as extensões nas outras dimensões.

-   Por exemplo, se `A` é uma matriz real da forma `(4,5,6)`, então o comando `SUM(A,DIM=2)` fornecerá uma matriz real `R` da forma `R(4,6)` cujo o elemento `(i,j)` tem o valor `R(i,j) = SUM(A(i,:,j))`.

    $$R(i,j) = \sum_{k}^{n} A_{i,k,j}$$

-   Como o rank da matriz resultante depende se a `DIM` é especificada, o
    argumento atual correspondente não deve ser opcionalmente um
    argumento mudo.

## Matrizes Automáticas

Certas situaçoes exigem que certo procedimento seja excutado, e toda vez que ele é chamado as matrizes que fazem parte de sua lista de argumentos, possuem diferentes tamanhos. Nesse caso, elas são alocadas automaticamente quando o procedimento é chamado. Um procedimento com argumentos matriciais mudos, pode ser necessário matrizes cujo tamanho varia de chamada para chamada ou ainda que também pode precisar de matrizes locais cujo tamanho varia. 

Pode-se dizer que as matrizes automáticas:

- São matrizes de forma explicita em um subprograma, cujos limites são fornecidos quando o subprograma é chamado via:
  - Argumentos mudos
  - Variáveis definidas pelo usuário ou por associação de hospedeiros
- Não devem aparecer em declarações `SAVE` e `NAMELIST` e nem devem ser inicializadas em declarações `TYPE`.

Note que o Fortran 90 não fornece nenhum mecanismo para verificar se existe memória suficiente para as matrizes automáticas. Se não houver, não há saída - o programa irá abortar.

Um exemplo simples é a matriz `mat` na subrotina para trocar duas matrizes que são mostradas no exemplo abaixo.

Uma matriz cujas dimensões/extensões variam dessa maneira é chamada de matriz automática e é um exemplo de um objeto de dados automático. Tal objeto não é um argumento mudo e sua declaração

```fortran
! Um procedimento com uma matriz automática. 
SUBROUTINE troca(a, b)
  IMPLICIT NONE
  REAL, DIMENSION(:), INTENT(inout) :: a, b
  REAL, DIMENSION(SIZE(a))          :: mat ! matriz automáticay
  ! a função intrínseca size fornece o tamanho da matriz
  mat = a
  a = b
  b = mat
END SUBROUTINE troca
```

contém um ou mais valores que não são conhecidos em tempo de compilação; isto é, não é uma expressão constante. Uma implementação pode trazê-los à existência quando o procedimento é chamado e destruí-los no retorno, mantendo-os em uma pilha. Uma **pilha** é um mecanismo de gerenciamento de memória pelo qual o armazenamento novo é estabelecido e o armazenamento antigo é descartado fora de base, muitas vezes dentro de memória contínua. Os valores devem ser definidos por expressões de especificação de matrizes.

A outra maneira que os objetos automáticos surgem é através do comprimento variável dos caracteres. A variável `texto2` no

```fortran
SUBROUTINE exemplo_texto(texto1)
  IMPLICIT NONE
  CHARACTER(len = *), INTENT(inout) :: texto1
  CHARACTER(len = LEN(texto1))      :: texto2
  :
END SUBROUTINE exemplo_texto

```

é um exemplo. Se um resultado de função tiver um comprimento de caractere variável, a interface deve ser explícita no ponto de chamada porque o compilador precisa saber disso, como mostra o exemplo anterior da subrotina `troca`.

```fortran

  CHARACTER(len = *), PARAMETER :: texto = 'Um teste simples'
  PRINT *, duplicar(texto)
CONTAINS
  FUNCTION duplicar(a)
    CHARACTER (len = *), INTENT(in) :: a
    CHARACTER (len = 2*LEN(a)+4)    :: duplicar
    duplicar = a// ' -> ' //a
  END FUNCTION duplicar
END PROGRAM teste_texto
```

O limite de uma matriz ou o comprimento do caractere de um objeto automático é fixado para a duração de cada execução do procedimento e não varia se o valor da expressão de especificação variar ou se tornar indefinido.

## Matrizes de forma assumida

Até aqui exigiu-se que as formas dos argumentos passados e mudos concordem, e até agora conseguimos isso passando as dimensões dos argumentos da matriz como argumentos adicionais. No entanto, é possível exigir que a forma da matriz muda seja automaticamente considerada como sendo o argumento da correspondente matriz passada. Tal matriz é dita ser uma matriz de forma assumida. Quando a forma é declarada pela cláusula de dimensão, cada dimensão tem o formulário

```fortran
[limite-inferior]:
```

na qual `limite-inferior` é uma expressão inteira que pode depender dos dados do módulo ou dos outros argumentos. Se o `limite-inferior` for omitido, o valor padrão no Fortran será 1. Observe que é a forma que é passada e não os limites superior e inferior. Por exemplo, se a matriz `mat` a ser passada for, declarada assim:

```fortran
REAL, DIMENSION (0:10, 0:20) :: mat
```

e se a matriz muda `mat_muda`,  for declarada assim:

```fortran
REAL, DIMENSION (:,:) :: mat_muda
```

então `mat(i,j)` corresponde a `mat_muda(i+1,j+1)`; para obter a correspondência natural, o limite inferior deve ser declarado:

```fortran
REAL, DIMENSION (0:, 0:) :: mat_muda
```

Para que o compilador saiba que informações adicionais devem ser fornecidas, a interface deve ser explícita (que será apresentada posteriormente) no ponto de chamada. Uma matriz muda com o atributo de ponteiro não é considerada como uma matriz de forma assumida porque sua forma não é necessariamente assumida.

Ao usar matrizes de forma assumida, em uma chamada a extensão real não é passada. Apenas o número de elementos é implicitamente passado. Se o limite inferior de uma extensão for diferente daquele do argumento real correspondente, a correspondência de elementos será diferente. É por isso que é importante que o subprogram saiba qual é o limite inferior correto da matriz que está sendo recebida por ele.
   
- Há três funções no Fortran 90 que podem ser usadas para recuperar a extensão e o tamanho de uma matriz:
   - `SIZE(M)` retorna o número de elementos na matrix M.
   - `LBOUND(M)` retorna o limite inferior da extensão da matriz M.
   - `UPPER(M)` retorna o limite superior da extensão da matriz M.

Se uma matriz de forma assumida `M()` tem um limite inferior correto, digamos `L`, então `SIZE(M)` retorna o número de elementos de `M()` e o limite superior de `M()` é `L + SIZE(M) - 1`. Portanto, para obter um limite superior correto, pode-se usar o seguinte truque:


```Fortran
SUBROUTINE Test_Mat(linf,Mat)
 IMPLICIT NONE
 INTEGER, INTENT(in) :: linf
 REAL, DIMENSION(linf:linf+SIZE(Mat)-1), INTENT(INOUT) :: Mat
 ...
END SUBROUTINE Test_Mat
```

Em síntese tem-se que

- A forma dos argumentos mudos e do atual devem concordar
- Fortran 77: passa as dimensões da matriz como argumentos
- Fortran 90: não passa necessariamente as dimensões de uma matriz
  - Matrizes de forma assumida, usam as dimensões dos argumentos atuais
  - Pode-se especificar o limite inferior
  - A necessidade de interfaces explicitas, devem fornecer um bloco de `INTERFACE` se usamos um subprograma externo

## Matrizes dinâmicas e alocação de  dados

No Fortran há uma condição subjacente de que o processador fornece um mecanismo para gerenciar o armazenamento da pilha. Lembre-se que a pilha é um mecanismo de gerenciamento de memória pelo qual o armazenamento recente pode ser estabelecido e o armazenamento antigo pode ser descartado em qualquer ordem. Mecanismos para lidar com a fragmentação progressiva da memória geralmente são necessários. As instruções descritas nesta seção são uma interface do usuário para esse mecanismo.

Por exemplo, há situações em que precisamos de uma matriz cujo o tamanho só será conhecido ao iniciarmos a execução do programa. No Fortran 77 para resolver esse problema, declarava-se a maior matriz possível, desse modo era possível rodar o programa sem ter problemas. Mas há situações em que a máquina que esta-se rodando o programa possui uma memória pequena, nesses caso nem sempre é possível rodar o programa. 

Atualmente, para permitir que o usuário consiga manipule a memória introduziu-se a partir do Fortran 90 a alocação dinâmica de matrizes.  

- No Fortran 77 tem-se somente:
  - Alocação estática de memória  
- A partir do Fortran 90 tem-se:
  - Alocação dinâmica de memória
  - Matrizes locais em um subprograma podem ter diferentes tamanhos e formas em cada instante que o subprograma é chamado via matrizes automáticas
  - Reduz a necessidade de armazenamento
  - Simplifica os argumentos das subrotinas


### O atributo  `ALLOCATABLE` 

Às vezes, é necessário que uma matriz tenha um tamanho conhecido apenas depois que alguns dados forem lidos ou alguns cálculos executados. Para este propósito, foi  introduzido no Fortran 90 o atributo `ALLOCATABLE` e as instruções `ALLOCATE` e  `DEALLOCATE`. Com isso, tornou-se possível definir a dimensão da matriz que tenha recebido o atributo `ALLOCATABLE` após iniciar o programa, usando o comando `ALLOCATE`. Nesse caso, a única informação que o fortran precisa saber, deve ser momento da definição da variável dando a ela o atributo `ALLOCATABLE`, o que significa que essa variável será alocada depois, conforme a declaração a seguir

```fortran
REAL, DIMENSION(:,:), ALLOCATABLE :: mat
```

Então a matriz `mat` é chamada de alocável, ou seja, tem o atributo `ALLOCATABLE`. Seu `RANK` deve ser especificado quando ela for declarada, mas os limites são indefinidos até uma que se tenha uma instrução de alocação.

Note que as matrizes que são declaradas com o atributo `ALLOCATABLE`, podem ser alocadas e desalocadas, o que libera memória, em qualquer momento durante a execução do programa, com so seguintes comandos:

- `ALLOCATE (allocate_object_list [, STAT= status])`, aloca a matriz
- `DEALLOCATE(allocate_object_list [, STAT= status])`, desaloca a matriz, liberando memória.
  - Quando `STAT=` está presente, `status = 0` (sucesso) o `status > 0` (erro).
  - Quando `STAT=` não está presente e um erro ocorre, o programa aborta a execução.


A seguinte instrução necessária para se alocar a matriz anterior,  


```fortran
ALLOCATE( mat(n, 0:n+1), stat=AllocStatus )  ! n é do tipo integer
```
Se o status de alocação, a variável `AllocStatus` for `0` a matriz foi alocada com sucesso, caso tenha outro valor houve problemas na alocação da matriz.


No exemplo a seguir, mostramos como compartilhas com outras partes do código, uma matriz dinâmica.  A matriz `mat` é colocada em um módulo e é alocada no início do programa principal para um tamanho que depende dos dados de entrada. A partir de então  a matriz estará disponível durante a execução do programa a qualquer subprograma que tenha a instrução `USE mat_temp`.

```fortran
! Uma matriz allocatable em um módulo.
MODULE mat_temp
  IMPLICIT NONE
  REAL, DIMENSION(:,:,:), ALLOCATABLE :: mat
END MODULE mat_temp

PROGRAM alocacao_dinamica
  USE mat_temp
  IMPLICIT NONE
  INTEGER :: nx, ny, nz
  
  ...
  
  READ *, nx, ny, nz
  ALLOCATE( mat(nx, ny, nz) )
  
  ...
  
END PROGRAM alocacao_dinamica
```

Quando uma matriz alocável `mat` não é mais necessária, ela pode ser desalocada com a seguinte instrução

```fortran
DEALLOCATE(mat)
```

a partir de então essa matriz não estará mais alocada. Use a instrução `DEALLOCATE` para liberar a memória.


Se for necessário fazer qualquer alteração nos limites de uma matriz alocável, a matriz deve ser desalocada e, em seguida, alocada novamente. Note que *essa restrição foi removida a partir do Fortran 2003*. É um erro tentar alocar uma matriz alocável que já está alocada ou desalocar uma matriz alocável que não está atualmente alocada, mas que pode ser facilmente evitada pelo uso da função intrínseca `ALLOCATED` que verifica o status de alocação de uma matriz. 

Um status de alocação indefinido não pode ocorrer. Ao retornar de um subprograma, uma matriz alocável alocada sem o atributo `SAVE`  é desalocado automaticamente se for local para o subprograma. Essa desalocação automática evita o vazamento inadvertido de memória.



###  Exemplos

Considere inicialmente o caso mais simples

```fortran
...
REAL, DIMENSION (:, :), ALLOCATABLE :: mat
INTEGER :: nl, nc, status
READ (*, *) nl, nc
ALLOCATE (mat(nl, nc), STAT = status)
IF (status > 0) STOP 'Falhou ao tentar alocamr Mat na memória'
...
! Antes de alocar, verifique se ela está alocada, e se 
! estiver alocada desaloque a matriz, para em seguida alocar.
IF (ALLOCATED(mat)) DEALLOCATE (mat)  
...

```


A seguinte forma para o código acima que envolvia um módulo, é mais recomendada


```fortran
! Uma matriz allocatable em um módulo.
MODULE mat_temp
  IMPLICIT NONE
  REAL, DIMENSION(:,:,:), ALLOCATABLE :: mat
CONTAINS
  SUBROUTINE alocar_mat(n1,n2,n3)
    IMPLICIT NONE
    INTEGER, INTENT (in) :: n1,n2,n3
    INTEGER :: aloca_status
    ! Verifique se ela já está alcoca, se sim desaloca
    IF (ALLOCATED (mat)) DEALLOCATE (mat)
    ! Aloca a matriz
    ALLOCATE( mat(n1,n2,n3), stat=aloca_status )
    ! O resultado da alocação
    IF( aloca_status /= 0 ) THEN
       PRINT*, ' ===> Problema na alocacao de memoria <=== '
       STOP 
    END IF
    RETURN
  END SUBROUTINE alocar_mat

  SUBROUTINE desalocar_mat()
    IMPLICIT NONE
    INTEGER :: erro, aloca_status
    ! Verifique se ela já está alcoca, se sim desaloca
    IF( ALLOCATED(mat) ) THEN
       DEALLOCATE (mat)
    ELSE
       PRINT *," A matriz não estava alocada"
    END IF
    RETURN
  END SUBROUTINE desalocar_mat

  ...

END MODULE mat_temp

PROGRAM alocacao_dinamica
  USE mat_temp
  IMPLICIT NONE
  INTEGER :: nx, ny, nz
  
  ...
  
  READ *, nx, ny, nz
  CALL alocar_mat(nx, ny, nz)
 
  ...

  CALL desaloca_mat
  
  ...
  
END PROGRAM alocacao_dinamica
```


A seguir apresentamos um exemplo de uso da alocação dinâmica de matrizes.

````python
%%writefile src/alocacao_dinamica_01.f95
PROGRAM Teste_de_Matrizes
  IMPLICIT NONE
  REAL, DIMENSION(:,:), ALLOCATABLE :: mat
  REAL    :: total, mem
  INTEGER :: dimensao, AllocateStatus, pr

  PRINT *, "Entre com a dimensão para a matriz quadrada: "
  READ *, dimensao

  ALLOCATE( mat(dimensao,dimensao), stat=AllocateStatus )

  IF( AllocateStatus /= 0 ) PRINT*, ' ===> Problema na alocacao de memoria <=== '

  PRINT *, ' A matriz foi alocada sim ou nao:' , ALLOCATED( mat )
  IF( ALLOCATED( mat ) ) PRINT *, 'Com o seguinte tamanho: ', SIZE( mat )
  
  pr = KIND(1.0)
  total =  REAL(dimensao) * REAL(dimensao)
  PRINT*
  PRINT*, 'A matriz tem o seguinte tamanho: ', total
  PRINT*, 'Em uma maquina de 64 bits, um real tem ', pr, 'Bytes'
  PRINT*, 'Para alocar sua matriz, voce precisou de uma memoria de'
  mem = pr * total
  IF ( mem < 1000.0 ) THEN 
     PRINT *, 'Memoria =', mem, 'bytes'
  ELSE IF (mem < 1000000.0 ) THEN 
     PRINT *, 'Memoria =', mem/1000.0, 'kbytes'
  ELSE IF (mem < 1000000000.0 ) THEN 
     PRINT *, 'Memoria =', mem/1000000.0, 'Mbytes'
  ELSE IF (mem < 1000000000000.0 ) THEN 
     PRINT *, 'Memoria =', mem/1000000000.0, 'Gbytes '    
  END IF 
END PROGRAM Teste_de_Matrizes
````
